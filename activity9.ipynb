## Task 5 Network Visualization with NetworkX


```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a random graph using Erdos-Renyi model
G = nx.erdos_renyi_graph(n=50, p=0.1, seed=42)

# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Normalize centrality values for coloring
colors = [degree_centrality[node] for node in G.nodes()]

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(
    G,
    with_labels=True,
    node_color=colors,
    cmap=plt.cm.viridis,
    node_size=500,
    edge_color='gray'
)
plt.title("Network Graph with Node Colors Based on Degree Centrality")
plt.show()
```

On the graph above, the nodes are colored based on their degree centrality, which measures the importance of a node in the network.\
The nodes with higher centrality are shown in brighter colors, indicating they are more connected.

## Task 6

In Visualization task, color selection is critical in visualization for clarity, aesthetics, and accessibility. it directly impacts how information is interpreted, understood, and acted upon. Poor color choices can obscure insights, mislead audiences, or make a visualization difficult to use. \
Key Concepts:\
- Color Contrast: Color contrast refers to the difference in lightness, saturation, or hue between two colors. It is essential for:
    Clarity: High contrast helps distinguish between elements, such as text and background, or between different data points in a chart.
    Readability: Ensuring text and graphical elements are legible, especially when printed or displayed on varying screen types.
    Highlighting Key Elements: Strategic contrast can draw attention to the most critical parts of the visualization.
- Color Harmony: Color harmony refers to the aesthetically pleasing arrangement of colors, often based on established color theories such as complementary, analogous, or triadic color schemes. Harmony ensures that colors:
Work Together: Avoid clashing and maintain a professional and cohesive appearance.
Enhance Communication: Make visualizations engaging without overwhelming the viewer.
- Accessibility ensures that the visualization can be understood by as many people as possible, including those with color vision deficiencies or other visual impairments. This includes:
Colorblind-Friendly Palettes: Avoid relying on red-green or blue-purple contrasts, which are problematic for many people with color blindness.
Redundant Encoding: Use shapes, patterns, or labels in addition to colors to convey information.
WCAG Compliance: The Web Content Accessibility Guidelines (WCAG) recommend a minimum contrast ratio of 4.5:1 for text and 3:1 for large text or graphical elements.\
-  Avoiding Overuse of Colors
Too many colors can confuse and overwhelm the audience. Simplify by:
Limiting the number of categories/colors to 5–7 for categorical data.
Using gradients or shades for sequential data (e.g., light to dark for increasing values).\
Examples:

Good Example: A heatmap using the "viridis" colormap, which is colorblind-friendly and perceptually uniform.
Bad Example: A pie chart with similar shades of red and green, making it indistinguishable for colorblind individuals.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample dataset
data = np.random.rand(10, 10)

# Good color palette
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(data, cmap="viridis", cbar=True)
plt.title("Good: Colorblind-Friendly Palette")

# Bad color palette
plt.subplot(1, 2, 2)
sns.heatmap(data, cmap="RdYlGn", cbar=True)
plt.title("Bad: Red-Green Palette")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load Titanic dataset
from seaborn import load_dataset
titanic = load_dataset('titanic')

# Clean data for visualization (drop NaN for age and fare)
titanic_cleaned = titanic.dropna(subset=['age', 'fare', 'sex', 'class', 'embarked'])

# Heatmap: Correlation matrix of numerical variables
correlation_matrix = titanic_cleaned[['age', 'fare', 'pclass', 'sibsp', 'parch']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Between Numerical Features in Titanic Dataset')
plt.show()

# Scatter Plot: Age vs Fare colored by survival and sized by passenger class
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=titanic_cleaned, 
    x='age', 
    y='fare', 
    hue='survived', 
    size='pclass', 
    sizes=(20, 200), 
    alpha=0.7
)
plt.title('Age vs Fare: Survival and Passenger Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()

# Insights from visualizations will follow this code block.
```

The correlation heatmap provides a high-level overview of relationships between numerical features in the Titanic dataset:

Fare and Pclass often exhibit a negative correlation, as higher classes typically have lower class numbers (e.g., 1st class is coded as 1).\
Survived shows moderate relationships with variables like Pclass and Fare, reflecting survival probabilities influenced by socio-economic status.\

Scatter Plot of Age vs. Fare:
This plot adds another layer by visualizing individual passenger data:
Age vs. Fare shows no clear linear relationship but clusters can be identified. For example, younger passengers might be concentrated in specific fare ranges, hinting at family groups or class preferences.

Together, the heatmap and scatter plot reveal both broad patterns and granular distributions. For instance, the negative correlation between Pclass and Fare is visually confirmed by clusters in the scatter plot.
